```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
df=sns.load_dataset('iris')
```

```python
df
```

```python
X=df.iloc[:,:-1]
y=dataset.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
#complete TREE

from sklearn.tree import DecisionTreeClassifier
```

```python
cls=DecisionTreeClassifier()
```

```python
cls.fit(X_train,y_train)
```

```python
cls.predict(X_test)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(12,10))
tree.plot_tree(cls,filled=True)
```

```python
#POST PRUNING
cls1=DecisionTreeClassifier(criterion='entropy',max_depth=2)
```

```python
cls1.fit(X_train,y_train)
```

```python
plt.figure(figsize=(5,6))
tree.plot_tree(cls1,filled=True)
```

```python
#PREDICTION
y_pred=cls1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#PRE PRUNING

parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']

}

```

```python
from sklearn.model_selection import GridSearchCV
```

```python
cv=GridSearchCV(cls,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
cv.best_score_
```

```python
y_pred=cv.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
