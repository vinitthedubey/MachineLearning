# Metrics Calculation

```python
# Calculate ROI for each account
df['ROI'] = df['realizedProfit'] / df['quantity']
```

```python
# Calculate total PnL for each account
pnl = df.groupby('Port_IDs')['realizedProfit'].sum().reset_index()
```

```python
list(pnl['realizedProfit'])
```

```python
# Calculate Sharpe Ratio for each account
returns = df.groupby('Port_IDs')['realizedProfit'].mean()
risk_free_rate = 0.01  # Example value
std_dev = df.groupby('Port_IDs')['realizedProfit'].std()
sharpe_ratio = (returns - risk_free_rate) / std_dev
```

```python
returns
```

```python
# Function to calculate Maximum Drawdown
def max_drawdown(returns):
    cumulative = (1 + returns).cumprod()
    peak = cumulative.cummax()
    drawdown = (cumulative - peak) / peak
    return drawdown.min()

# Calculate MDD for each account
df['returns'] = df['realizedProfit'] / df['quantity']
mdd = df.groupby('Port_IDs')['returns'].apply(max_drawdown).reset_index()
```

```python
list(mdd['returns'])
```

```python
# Calculate Win Rate for each account
win_rate = df[df['realizedProfit'] > 0].groupby('Port_IDs').size() / df.groupby('Port_IDs').size()
```

```python
list(win_rate)
```

```python
# Count Win Positions for each account
win_positions = df[df['realizedProfit'] > 0].groupby('Port_IDs').size().reset_index(name='win_positions')
```

```python
win_positions
```

```python
# Count Total Positions for each account
total_positions = df.groupby('Port_IDs').size().reset_index(name='total_positions')
```

```python
metrics_df = pd.DataFrame({
    'Port_ID': list(df['Port_IDs'].unique()),
    'ROI': (list(df['ROI'].groupby(df['Port_IDs']).mean())),
    'PnL': list(pnl['realizedProfit']),
    'Sharpe_Ratio': list(sharpe_ratio),
    'MDD': list(mdd['returns']),
    'Win_Rate': list(win_rate),
    'Win_Positions': list(win_positions['win_positions']),
    'Total_Positions': list(total_positions['total_positions'])
})
```

```python
# Define weights for each metric
weights = {
    'ROI': 0.2,
    'PnL': 0.2,
    'Sharpe_Ratio': 0.2,
    'MDD': -0.2,  # Negative weight for MDD as it is a risk measure
    'Win_Rate': 0.1,
    'Win_Positions': 0.05,
    'Total_Positions': 0.05
}
```

```python
# Calculate weighted score for each account
metrics_df['score'] = (
    weights['ROI'] * metrics_df['ROI'] +
    weights['PnL'] * metrics_df['PnL'] +
    weights['Sharpe_Ratio'] * metrics_df['Sharpe_Ratio'] +
    weights['MDD'] * metrics_df['MDD'] +
    weights['Win_Rate'] * metrics_df['Win_Rate'] +
    weights['Win_Positions'] * metrics_df['Win_Positions'] +
    weights['Total_Positions'] * metrics_df['Total_Positions']
)
```

```python
# Rank accounts based on the score
metrics_df['rank'] = metrics_df['score'].rank(ascending=False)
top_20_accounts = metrics_df.nsmallest(20, 'rank')
```

```python
# Display the top 20 accounts
print(top_20_accounts)
```

```python
# Save the top 20 accounts to a CSV file
top_20_accounts.to_csv('Final_prediction.csv', index=False)
```
