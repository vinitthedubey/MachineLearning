## DBSCAN Clutering

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y=make_moons(n_samples=250,noise=0.10)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
##feature scaling(Standard Scaling)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
dbcan=DBSCAN(eps=0.3)
```

```python
dbcan.fit(X_scaled)
```

```python
dbcan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbcan.labels_)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
## silhouette score
from sklearn.metrics import silhouette_score
```

```python
scaled_data=X_scaled
silhouette_coefficients = []

# Notice you start at 2 clusters for silhouette coefficient
k=0.1
while(k<10):
    agglo = DBSCAN(eps=k)
    agglo.fit(scaled_data)
    score = silhouette_score(scaled_data,labels=dbcan.labels_)
    silhouette_coefficients.append(score)
    k+=0.3
```

```python
dbcan.labels_
```

```python
silhouette_coefficients
```

```python
## plotting silhouette score
plt.plot(silhouette_coefficients)
plt.xlabel("Number of Cluters")
plt.ylabel("Silhoutte Coeffecient")
plt.show()
```
